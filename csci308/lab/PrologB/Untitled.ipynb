```python
def to_lean_json(l):
    if type(l) == list:
        if len(l) == 0:
            return "json.array ([])"
        return "json.array (" + "[" + ", ".join([to_lean_json(x) for x in l if x != ""]) + "]" + ")"
    elif type(l) == str:
        return f'json.of_string "{l}"'
    elif type(l) == bool:
        return "json.of_bool ff" if False else "json.of_bool tt"
    elif type(l) == int:
        return f"json.of_int {l}"
    else:
        raise Exception(f"Found object {l} of type {type(l)}")
```

```python
def get_len(l):
    if type(l) != list or len(l) == 0:
        return None
    
    ret = []
    for x in l:
        temp = get_len(x)
        if temp != None:
            ret.append(temp)
        else:
            ret.append("")
    ret.append(len(ret))
    return ret
```
